I will first create a new column called 'max_value' that will indiacte the customers total orders. Then create another column called loyalty flag to indicate whether the customer is Loyal - 40 orders and above, Regular - in between 10 and 40 orders and new customer less than 10 orders:

```python
# create 'max_order' column by grouping 'user_id' to determine amount of orders that each customer places
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

# Question 7

I will first create a new column called 'customer frequency'. I will group the data by 'user_id' and calculate the median days of öast order by finding the median of the 'days_since_prior_order' column for each 'user_id'.  Then I will create a new column customer_frequency to highlight the status of the customer.


Non-frequent customer = more than 20 days since last order

Regular customer = higher than 10 but less than or equal to 20  days since last order

Frequent customer = less than or equal to 10 days from last order

```python
# Create a new column for customer frequency flags based on the median values from 'days_since_prior_order' column
df_opm['customer_frequency'] = df_opm.groupby(['user_id'])['days_since_last_order'].transform('median')
```

```python
# Check that the column was created successfully and properly
df_opm.head(20)
```

```python
# Based on the criteria provided, create a column featuring order frequency flags
df_opm.loc[df_opm['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_opm.loc[(df_opm['customer_frequency'] > 10) & (df_opm['customer_frequency'] <=20), 'frequency_flag'] = 'Regular customer'
```

```python
df_opm.loc[df_opm['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check value counts for new column
df_opm['frequency_flag'].value_counts(dropna = False)
```
